```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the sampling rate and duration of the signal
fs = 1000 # Hz
duration = 60 # seconds
```

```python
# Generate an artificial HRV signal for distress condition
mean_hr_distress = 80 # bpm
std_hr_distress = 15 # bpm
mean_rr_distress = 60 / mean_hr_distress # seconds
std_rr_distress = std_hr_distress / mean_hr_distress * mean_rr_distress # seconds
num_samples = duration * fs // 2
rr_intervals_distress = np.random.normal(mean_rr_distress, std_rr_distress, size=num_samples)
timestamps_distress = np.cumsum(rr_intervals_distress)
heart_rate_distress = 60 / rr_intervals_distress
```

```python
# Generate an artificial HRV signal for eustress condition
mean_hr_eustress = 60 # bpm
std_hr_eustress = 10 # bpm
mean_rr_eustress = 60 / mean_hr_eustress # seconds
std_rr_eustress = std_hr_eustress / mean_hr_eustress * mean_rr_eustress # seconds
num_samples = duration * fs // 2
rr_intervals_eustress = np.random.normal(mean_rr_eustress, std_rr_eustress, size=num_samples)
timestamps_eustress = np.cumsum(rr_intervals_eustress)
heart_rate_eustress = 60 / rr_intervals_eustress
```

```python
# Combine the signals into one signal
timestamps = np.concatenate((timestamps_distress, timestamps_eustress + timestamps_distress[-1]))
heart_rate = np.concatenate((heart_rate_distress, heart_rate_eustress))
```

```python
# Plot the HRV signal
plt.figure()
plt.plot(timestamps, heart_rate)
plt.xlabel('Time (s)')
plt.ylabel('Heart rate (bpm)')
plt.show()
```
