```python
import pandas as pd
import numpy as np
# import pyhrv

# Function to calculate HRV measures for a segment
def calculate_hrv_measures(rr_segment):
    # Convert RR interval data to a format suitable for pyHRV, if necessary
    # Calculate HRV measures using pyHRV or other methods
    # For demonstration, let's assume we calculate these three measures
    return {
        'SDNN': np.std(rr_segment),  # Standard deviation
        'RMSSD': np.sqrt(np.mean(np.square(np.diff(rr_segment)))),  # RMSSD
        'MeanRR': np.mean(rr_segment)  # Mean RR interval
    }

file_path = folder_path + '/000.txt'
# Read RR interval data from file, ignoring empty lines or non-integer values
with open(file_path, 'r') as file:
    rr_intervals = []
    for line in file:
        stripped_line = line.strip()
        if stripped_line.isdigit():  # Check if the line contains digits only
            rr_intervals.append(int(stripped_line))

# Segment the data into one-minute chunks
# Assuming a constant heart rate of 60 beats per minute, which is simplistic
# In reality, the segmentation should consider the actual timestamps
one_minute_segment_length = 60  # Adjust based on your data's actual heart rate
segments = [rr_intervals[i:i + one_minute_segment_length] for i in range(0, len(rr_intervals), one_minute_segment_length)]

# Calculate HRV measures for each segment
hrv_measures_per_minute = [calculate_hrv_measures(segment) for segment in segments if len(segment) == one_minute_segment_length]

# Convert the results to a pandas DataFrame
hrv_df = pd.DataFrame(hrv_measures_per_minute)

# Write the DataFrame to a CSV file
hrv_df.to_csv(folder_path + '/000.csv', index=False)

print(f"HRV measures per minute have been saved.")
```
