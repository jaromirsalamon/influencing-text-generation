```python
import numpy as np

# Setting the seed for reproducibility
np.random.seed(0)

def generate_binary_data(duration):
  time = np.arange(duration)  # Time in minutes
  binary_signal = np.random.randint(2, size=duration)

  return time, binary_signal

time, binary_stress = generate_binary_data(1000)
```

```python
# Calculating the ratio of '0's (No Stress) to '1's (Stress) in the 1000 samples
no_stress_count = np.count_nonzero(binary_stress == 0)
stress_count = np.count_nonzero(binary_stress == 1)

# Ratio calculation
no_stress_ratio = no_stress_count / len(binary_stress)
stress_ratio = stress_count / len(binary_stress)

no_stress_ratio, stress_ratio
```

```python
# Adding counts to the bars in the bar chart

values = [0, 1]
counts = [no_stress_count, stress_count]

# Recreating the bar chart
plt.figure(figsize=(12, 6))
plt.bar(values, counts, color=['blue', 'orange'])
plt.title('Distribution of Binary Stress Data', fontsize=16)
plt.ylabel('Counts', fontsize=14)
plt.xticks(values, ['No Stress (0)', 'Stress (1)'], fontsize=12)
plt.yticks(fontsize=12)
# plt.grid()

# Adding the counts on top of each bar
for i in range(len(counts)):
    plt.text(i, 510, str(counts[i]), ha='center', fontsize=12)

# Show the plot with numbers on bars
plt.savefig(img_path + 'distribution_binary_stress_data.png', dpi=300, bbox_inches='tight')
plt.show()
```
