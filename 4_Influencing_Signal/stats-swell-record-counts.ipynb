```python
import pandas as pd
import glob

# Pattern to match files, assuming the current directory. Adjust the path as necessary.

file_list = [f"p{i}_RR.csv" for i in range(1, 25)]

total_records = 0

for file_name in file_list:
    try:
        df = pd.read_csv(folder_input_path + file_name)
        total_records += len(df)
        print(f"File {file_name} has {len(df)} records.")
    except Exception as e:
        print(f"Error reading {file_name}: {e}")

print(f"Total number of records across all files: {total_records}")
```
