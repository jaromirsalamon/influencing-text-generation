```python
from bert_score import score

def compute_bertscore(original, paraphrase):
    # Since BERTScore expects lists of sentences, wrap inputs in lists
    P, R, F1 = score([paraphrase], [original], lang="en", verbose=False)

    # Return the F1 score
    return F1.item()
```

```python
original = "Our team has bearable performance base."
positive = "Our team is well-positioned for a triumphant victory."
negative = "Our team's chances of winning are quite doubtful."
```

```python
%%time

compute_bertscore(original, positive)
```

```python
%%time

compute_bertscore(original, negative)
```
