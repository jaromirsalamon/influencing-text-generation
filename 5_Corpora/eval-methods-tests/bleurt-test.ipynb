```python
from bleurt import score

# Load the BLEURT scorer with the path to your downloaded model
bleurt_scorer = score.BleurtScorer("/content/drive/My Drive/Colab Notebooks/Ch3_Corpora/bleurt-model")

def compute_bleurt_score(original, paraphrase):
    scores = bleurt_scorer.score(references=[original], candidates=[paraphrase])
    return scores[0]

# Example usage
reference = "The quick brown fox jumps over the lazy dog."
candidate = "A quick brown fox leaped over the lazy dog."
bleurt_score = compute_bleurt_score(reference, candidate)
print(f"BLEURT Score: {bleurt_score}")
```
