```python
import sacrebleu

def compute_chrf(original, paraphrase, beta=2, word_order=False):
    # Compute chrF and optionally chrF++ if word_order is True
    chrf_score = sacrebleu.corpus_chrf(paraphrase, [original], beta=beta, word_order=word_order).score

    return chrf_score
```

```python
def compute_ter(original, paraphrase):
    ter = sacrebleu.metrics.TER()
    score = ter.sentence_score(paraphrase, [original])
    return score.score
```

```python
original = "Our team has bearable performance base."
positive = "Our team is well-positioned for a triumphant victory."
negative = "Our team's chances of winning are quite doubtful."
```

```python
%%time

chrf_score = compute_chrf(original, positive)
chrfpp_score = compute_chrf(original, positive, word_order=True)

print(f"chrF score: {chrf_score}")
print(f"chrF++ score: {chrfpp_score}")
```

```python
%%time

chrf_score = compute_chrf(original, negative)
chrfpp_score = compute_chrf(original, negative, word_order=True)

print(f"chrF score: {chrf_score}")
print(f"chrF++ score: {chrfpp_score}")
```

```python
positive_ter_score = compute_ter(original, positive)
negative_ter_score = compute_ter(original, negative)

print(f"positive TER score: {positive_ter_score}")
print(f"negative TER score: {negative_ter_score}")
```
