```python
from rouge_score import rouge_scorer

def calculate_rouge_f1_scores(original, paraphrase):
    # Create a ROUGE scorer instance
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

    # Calculate ROUGE scores for the paraphrase
    scores = scorer.score(original, paraphrase)

    # Extract and return the F1-scores for ROUGE-1, ROUGE-2, and ROUGE-L
    rouge1_f1 = scores['rouge1'].fmeasure
    rouge2_f1 = scores['rouge2'].fmeasure
    rougeL_f1 = scores['rougeL'].fmeasure

    return rouge1_f1, rouge2_f1, rougeL_f1
```

```python
original = "Our team has bearable performance base."
positive = "Our team is well-positioned for a triumphant victory."
negative = "Our team's chances of winning are quite doubtful."
```

```python
calculate_rouge_f1_scores(original, positive)
```

```python
calculate_rouge_f1_scores(original, negative)
```
