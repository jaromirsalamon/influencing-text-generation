```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Define cosine similarities for the specific relationships
cosine_sim_positive = 0.8  # Original to Positive Paraphrase
cosine_sim_negative = 0.6  # Original to Negative Paraphrase

# Sentiment changes: neutral to negative (-1), and neutral to positive (+1)
sentiments = {'Negative': -1, 'Positive': 1}
```

```python
# Cosine similarities for demonstration, represented as angles
# Note: Here, we're interpreting the cosine similarity in a conceptual way to fit the angle representation
cosine_similarity_to_angle = {
    'Negative': 90 * cosine_sim_negative,  # Example similarity for negative sentiment
    'Positive': 90 * cosine_sim_positive,  # Example similarity for positive sentiment
}

# Create figure and axis for plotting
fig, ax = plt.subplots(figsize=(8, 5))

# Plot vectors representing sentiment changes with cosine similarity
for sentiment, x_value in sentiments.items():
    angle = cosine_similarity_to_angle[sentiment]
    radians = np.radians(angle)
    y_value = np.sin(radians)  # Y-value based on cosine similarity angle
    ax.quiver(0, 0, x_value, y_value, angles='xy', scale_units='xy', scale=1, label=f'{sentiment} (sim={angle}°)')

# Configure plot
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(0, 1.5)
ax.set_xlabel('Sentiment Shift')
ax.set_ylabel('Cosine Similarity Angle Representation')
ax.set_title('Corrected Sentiment Shift and Cosine Similarity Visualization')
ax.grid(True)
plt.axhline(0, color='black', linewidth=0.5)  # Neutral sentiment line
plt.legend()
plt.show()
```

```python
# Points for positive and negative sentiment vectors based on cosine similarity
positive_vector_A = (0, 0)  # Starting point
positive_vector_B = (sentiments['Positive'], cosine_sim_positive)  # Ending point based on cosine similarity

negative_vector_A = (0, 0)  # Starting point
negative_vector_B = (sentiments['Negative'], cosine_sim_negative)  # Ending point based on cosine similarity

# Create figure and axis for plotting
fig, ax = plt.subplots(figsize=(8, 5))

# Adding vectors with arrows for positive and negative sentiments
# Positive sentiment vector with an arrow
ax.quiver(positive_vector_A[0], positive_vector_A[1], positive_vector_B[0]-positive_vector_A[0], positive_vector_B[1]-positive_vector_A[1],
          angles='xy', scale_units='xy', scale=1, color='green', label='Positive Sentiment')

# Negative sentiment vector with an arrow
ax.quiver(negative_vector_A[0], negative_vector_A[1], negative_vector_B[0]-negative_vector_A[0], negative_vector_B[1]-negative_vector_A[1],
          angles='xy', scale_units='xy', scale=1, color='red', label='Negative Sentiment')

# Configure plot
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(0, 1)
ax.set_xlabel('Sentiment Shift')
ax.set_ylabel('Cosine Similarity Value')
ax.set_title('Visualization of Sentiment Shift and Cosine Similarity with Direction')
ax.grid(True)
plt.axhline(0, color='black', linewidth=0.5)  # Neutral sentiment line
plt.legend()

plt.savefig(img_path + 'triplets-sentiment-shift.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dummy values for demonstration purposes
cosine_sim_positive = 0.8
cosine_sim_negative = 0.6
sentiments = {'Positive': 1, 'Negative': -1}

# Points for vectors
positive_vector_A = (0, 0)  # Starting point
positive_vector_B = (sentiments['Positive'], cosine_sim_positive)  # Positive sentiment

negative_vector_A = (0, 0)  # Starting point
negative_vector_B = (sentiments['Negative'], cosine_sim_negative)  # Negative sentiment

# Create figure and axis for plotting
fig, ax = plt.subplots(figsize=(12, 6))

# Adding vectors with arrows for positive and negative sentiments
# Positive sentiment vector with an arrow
ax.quiver(positive_vector_A[0], positive_vector_A[1], positive_vector_B[0]-positive_vector_A[0], positive_vector_B[1]-positive_vector_A[1],
          angles='xy', scale_units='xy', scale=1, color='green', label='Positive sentiment')

# Negative sentiment vector with an arrow
ax.quiver(negative_vector_A[0], negative_vector_A[1], negative_vector_B[0]-negative_vector_A[0], negative_vector_B[1]-negative_vector_A[1],
          angles='xy', scale_units='xy', scale=1, color='red', label='Negative sentiment')

# Points and annotations
points = {'Original sentence': (0, 0), 'Negative sentence': (-1, 0.6), 'Positive sentence': (1, 0.8)}
for label, (x, y) in points.items():
    ax.plot(x, y, 'o', markersize=7)  # Plot point
    ax.text(x + 0.1, y + 0.02, label) if label == 'Original sentence' else None # Annotate point
    ax.text(x - 0.2, y + 0.03, label) if label == 'Negative sentence' else None
    ax.text(x - 0.2, y + 0.03, label) if label == 'Positive sentence' else None

# Configure plot
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(0, 1)

# Set x-axis to be discrete categorical
ax.set_xticks([-1, 0, 1])
ax.set_xticklabels(['Negative sentiment', 'Neutral sentiment', 'Positive sentiment'], fontsize=14)


ax.set_ylabel('Sentence similarity value', fontsize=14)
ax.set_title('Sentence sentiment shift and similarity representation', fontsize=16)
ax.grid(True)
plt.axhline(0, color='black', linewidth=0.5)  # Neutral sentiment line
plt.legend(fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Replace 'img_path' with the actual path where you want to save the image
plt.savefig(img_path +'triplets-sentiment-shift.png', dpi=300, bbox_inches='tight')
plt.show()
```
