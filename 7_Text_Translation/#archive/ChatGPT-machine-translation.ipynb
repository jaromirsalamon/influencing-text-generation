```python
from transformers import MarianMTModel, MarianTokenizer

def translate(text, src_lang, tgt_lang):
    # Define the model name
    model_name = f'Helsinki-NLP/opus-mt-{src_lang}-{tgt_lang}'

    # Load tokenizer and model from pretrained MarianMT
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    model = MarianMTModel.from_pretrained(model_name)

    # Tokenize the text
    translated = model.generate(**tokenizer(text, return_tensors="pt", padding=True))

    # Decode the translated text
    translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)

    return translated_text

# Example usage
source_text = "Hello, world!"
source_language = "en"  # English
target_language = "fr"  # French

translated_text = translate(source_text, source_language, target_language)
print(f"Translated text: {translated_text}")
```

```python
from langdetect import detect

def detect_language(text):
    try:
        # Detect the language
        language = detect(text)
        return language
    except Exception as e:
        return f"Error in language detection: {e}"

# Example usage
# text = "Bonjour le monde"
detected_language = detect_language(translated_text)
print(f"Detected language: {detected_language}")
```
