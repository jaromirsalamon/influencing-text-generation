```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
```

```python
model_name = "Vamsi/T5_Paraphrase_Paws"  # This is a T5 model fine-tuned for paraphrasing
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
def paraphrase(text):
    # Encode the input text and convert it to tensor
    input_ids = tokenizer.encode(text, return_tensors="pt", add_special_tokens=True)

    # Generate paraphrase
    paraphrased_ids = model.generate(input_ids, num_beams=5, no_repeat_ngram_size=2, num_return_sequences=1, max_length=50)
    paraphrased_text = tokenizer.decode(paraphrased_ids[0], skip_special_tokens=True)

    return paraphrased_text
```

```python
original_text = "It's a challenging task."
paraphrased_text = paraphrase(original_text)
print("Original:", original_text)
print("Paraphrased:", paraphrased_text)
```
