```python
import pandas as pd
from transformers import pipeline
from tqdm import tqdm
tqdm.pandas()

# Load the sentiment analysis pipeline
model = pipeline('sentiment-analysis', model="finiteautomata/bertweet-base-sentiment-analysis")

# Function to get sentiment label and score
def get_sentiment(sentence):
    result = model(sentence)[0]
    return result['label'], result['score']

# Apply the function to each sentence column and create new columns for label and score
for column in ['sentence_orig', 'paraphrase_pos', 'paraprhase_neg']:
    df[[column + '_label', column + '_score']] = df[column].progress_apply(
        lambda x: pd.Series(get_sentiment(x))
    )
```

```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd

# Load the model
model = SentenceTransformer('all-mpnet-base-v2')

def is_paraphrase(sentence1, sentence2, threshold=0.75):
    # Compute embeddings
    embeddings1 = model.encode(sentence1, convert_to_tensor=True)
    embeddings2 = model.encode(sentence2, convert_to_tensor=True)

    # Compute cosine similarity
    cosine_scores = util.pytorch_cos_sim(embeddings1, embeddings2)

    return cosine_scores.item() # >= threshold

# Assuming your DataFrame 'df' has columns 'sentence_orig', 'paraphrase_pos', 'paraphrase_neg'
for index, row in df.iterrows():
    df.at[index, 'pos_is_paraphrase'] = is_paraphrase(row['sentence_orig'], row['paraphrase_pos'])
    df.at[index, 'neg_is_paraphrase'] = is_paraphrase(row['sentence_orig'], row['paraprhase_neg'])

# df now includes whether each paraphrase is considered a paraphrase of the original sentence
```
